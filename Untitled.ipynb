```python
from imblearn.combine import SMOTETomek
```

```python
acc=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,temp,y)#defauly use statified cross val
    acc.append(score.mean())
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),acc,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('accuracy vs Kval')
plt.xlabel('K')
plt.ylabel('accyracy Rate')
```
